```python
import pandas as pd
from telegram import Update
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext

# Replace 'YOUR_TOKEN' with your actual bot token
TOKEN = '7266191209:AAHVd8RKUd-j0EeDdRdsas5PlJfkvbn6JnU'
EXCEL_FILE = 'trades.xlsx'

def read_excel():
    return pd.read_excel(EXCEL_FILE)

def write_excel(df):
    df.to_excel(EXCEL_FILE, index=False)

def dca_price(existing_price, existing_size, new_price, new_size):    #size@price
    total_cost = (existing_price * existing_size) + (new_price * new_size)
    total_size = existing_size + new_size
    return total_cost / total_size, total_size

def update_trade(update: Update, context: CallbackContext) -> None:
    args = context.args
    if len(args) != 3:
        update.message.reply_text('Usage: /update <stock> <new_price> <new_size>')
        return

    stock, new_price, new_size = args
    new_price = float(new_price)
    new_size = int(new_size)

    df = read_excel()

    if stock in df['Stock'].values:
        row = df[df['Stock'] == stock].iloc[0]
        existing_price = row['Price']
        existing_size = row['Lot Size']

        updated_price, updated_size = dca_price(existing_price, existing_size, new_price, new_size)

        df.loc[df['Stock'] == stock, 'Price'] = updated_price
        df.loc[df['Stock'] == stock, 'Lot Size'] = updated_size
    else:
        df = df.append({'Stock': stock, 'Price': new_price, 'Lot Size': new_size}, ignore_index=True)

    write_excel(df)
    update.message.reply_text(f'Trade updated for {stock}. New Price: {updated_price:.2f}, New Size: {updated_size}')

def main():
    # Create the Updater and pass it your bot's token
    updater = Updater(TOKEN)

    # Get the dispatcher to register handlers
    dispatcher = updater.dispatcher

    # Register the /update command handler
    dispatcher.add_handler(CommandHandler("update", update_trade))

    # Start the Bot
    updater.start_polling()

    # Run the bot until you press Ctrl+C or the process receives SIGINT, SIGTERM or SIGABRT
    updater.idle()

if __name__ == '__main__':
    main()
```
